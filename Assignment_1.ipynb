This function uses a dictionary to store numbers and their indices. For each element, it checks if target - num already exists in the dictionary. If yes, the pair is found. This solution runs in O(n) time.

````python
# Q1. Given a list of integers “nums” and an integer “target”, return indices of the two numbers that add up to the target.

def two_sum(nums, target):
    lookup = {}
    for i, num in enumerate(nums):
        if target - num in lookup:
            return [lookup[target - num], i]
        lookup[num] = i

# Example test cases
print(two_sum([2,7,11,15], 9))   # Output: [0, 1]
print(two_sum([3,2,4], 6))       # Output: [1, 2]
print(two_sum([3,3], 6))         # Output: [0, 1]
````

The number is converted to a string and compared to its reverse (s[::-1]). If both are equal, it’s a palindrome. Negative numbers fail this check automatically.

````python
# Q2. Check if an integer is a palindrome.

def is_palindrome(x):
    s = str(x)
    return s == s[::-1]

# Example test cases
print(is_palindrome(121))   # Output: True
print(is_palindrome(-121))  # Output: False
print(is_palindrome(10))    # Output: False
````

Start with the first string as a prefix and shorten it until all words begin with it. If no common part remains, return an empty string.

````python
# Q3. Find the longest common prefix string amongst an array of strings.

def longest_common_prefix(strs):
    if not strs:
        return ""
    prefix = strs[0]
    for s in strs[1:]:
        while not s.startswith(prefix):
            prefix = prefix[:-1]
            if prefix == "":
                return ""
    return prefix

# Example test cases
print(longest_common_prefix(["flower","flow","flight"]))  # Output: "fl"
print(longest_common_prefix(["dog","racecar","car"]))     # Output: ""
````

Roman numerals subtract smaller values before larger ones (e.g., IV = 4). The code checks this by comparing each numeral to the next and adding or subtracting accordingly.

````python
# Q4. Convert Roman numerals into integers.

def roman_to_int(s):
    roman = {'I':1, 'V':5, 'X':10, 'L':50, 'C':100, 'D':500, 'M':1000}
    total = 0
    for i in range(len(s)):
        if i + 1 < len(s) and roman[s[i]] < roman[s[i+1]]:
            total -= roman[s[i]]
        else:
            total += roman[s[i]]
    return total

# Example test cases
print(roman_to_int("III"))      # Output: 3
print(roman_to_int("LVIII"))    # Output: 58
print(roman_to_int("MCMXCIV"))  # Output: 1994
````

Uses binary search to find the square root. If mid*mid is less than x, shift left; otherwise, move right. The integer part of the square root is returned.

````python
# Q5. Compute the square root of a number without using built-in sqrt, pow, or **.

def my_sqrt(x):
    if x < 2:
        return x
    left, right = 1, x // 2
    while left <= right:
        mid = (left + right) // 2
        if mid * mid == x:
            return mid
        elif mid * mid < x:
            left = mid + 1
        else:
            right = mid - 1
    return right

# Example test cases
print(my_sqrt(4))  # Output: 2
print(my_sqrt(8))  # Output: 2
````
